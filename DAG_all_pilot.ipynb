```python
normalization = math.sqrt(math.log(2))
```

```python
network_file_CB = "DAG_3rd_pilot_postQ.xdsl"
network_file_CT = "DAG_CT_v2_trained.xdsl"
network_file_FN = "DAG_FN_v2_trained.xdsl"
network_file_OG = "DAG_OG_v2_trained.xdsl"
```

## FN / no FN (Top -> Down)

```python
var_interest = 'FN_risk'
```

```python
net = pysmile.Network()
net.read_file(network_file_FN)
# net.set_evidence("Honesty", 3)
net.update_beliefs()

lst_var_FN = ['Age',
 'Gender',
 'Family_Support',
 'Social_Support',
 'Hours_Internet',
 'CB_Offending',
 'Migratory_background',
 'Conscientiousness_BF',
 'Agreeableness_BF',
 'Neuroticism_BF',
 'FN_risk',
 'Honesty',
 'A6_6_regarding_charles',
 'A6_4_accu',
 'A6_5_replay_post',
 'A6_3_source',
 'A6_2_look',
 'A6_1_news_check']

# CB_Offending = None
net.set_evidence(var_interest, 0)
net.update_beliefs()
dict_beliefs_0 = {}
for var in lst_var_FN:
  beliefs = net.get_node_value(var)
  dict_beliefs_0[var] = beliefs

# CB_Offending = Bully
net.set_evidence(var_interest, 1)
net.update_beliefs()
dict_beliefs_1 = {}
for var in lst_var_FN:
  beliefs = net.get_node_value(var)
  dict_beliefs_1[var] = beliefs

# KL divergence
def kl_divergence(p,q):
	return sum((p[i] * math.log(p[i]/q[i]) for i in range(len(p))))

# Jensen-Shannon distances
dir_distances = {}
for var in lst_var_FN:
  arr_0 = dict_beliefs_0.get(var)
  arr_1 = dict_beliefs_1.get(var)
  dist = scipy.spatial.distance.jensenshannon(arr_0, arr_1)/normalization
  # dist = kl_divergence(arr_0, arr_1)
  dir_distances[var] = dist

# print sorted results
sorted(dir_distances.items(), key=lambda x: -x[1])
```
